```python
#Eresion and Dilation
from re import I
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/drive/MyDrive/ColabPhotos/morp.png',0)
print("Normal Image: ")
cv2_imshow(img)

kernel = np.ones((4,4),np.uint8)

image_erosion = cv2.erode(img, kernel, iterations=1)
image_dilation = cv2.dilate(img, kernel, iterations=1)

print("Erosion Image: ")
cv2_imshow(image_erosion)

print("Dilation Image: ")
cv2_imshow(image_dilation)
```

```python
#Opening
img = cv2.imread('/content/drive/MyDrive/ColabPhotos/omerSenol.png',0)
print("Normal Image: ")
cv2_imshow(img)

kernel = np.ones((4,4),dtype = np.uint8)

whiteNoise = np.random.randint(0,2,size=img.shape[:2])
whiteNoise = whiteNoise*255

noise_img = whiteNoise + img

print("Noisy Image: ")
cv2_imshow(noise_img)

opening = cv2.morphologyEx(noise_img.astype(np.float32),cv2.MORPH_OPEN,kernel)
print("Opening Image: ")
cv2_imshow(opening)
```

```python
#Closing
img = cv2.imread('/content/drive/MyDrive/ColabPhotos/omerSenol.png',0)
print("Normal Image: ")
cv2_imshow(img)

kernel = np.ones((5,5),dtype = np.uint8)

blackNoise = np.random.randint(0,2,size=img.shape[:2])
blackNoise = blackNoise * -255
noise_img = blackNoise + img

noise_img[noise_img <= -245] = 0

print("Noisy Image: ")
cv2_imshow(noise_img)

closing = cv2.morphologyEx(noise_img.astype(np.float32),cv2.MORPH_CLOSE,kernel)
print("Closing Image: ")
cv2_imshow(closing)
```
