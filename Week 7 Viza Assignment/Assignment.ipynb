```python
#Gamma Transform
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/drive/MyDrive/ColabPhotos/bird.jpg')

gamma_3 = np.array(255*(img/255)**3,dtype='uint8')

gamma_4 = np.array(255*(img/255)**4,dtype='uint8')

gamma_5 = np.array(255*(img/255)**5,dtype='uint8')

gamma_0_2 = np.array(255*(img/255)**0.2,dtype='uint8')

#Normal Image
print("Normal Image:")
cv2_imshow(img)

#Gamma Images
print("Gamma 3.0:")
cv2_imshow(gamma_3)
print("Gamma 4.0:")
cv2_imshow(gamma_4)
print("Gamma 5.0:")
cv2_imshow(gamma_5)

#Addtionaly I add 0.2 gamma for the see brightness
print("Gamma 0.2:")
cv2_imshow(gamma_0_2)
    
```

```python
#Log Transformation

img = cv2.imread('/content/drive/MyDrive/ColabPhotos/womanFace.jpg',0)

#Apply log transform
img_log = (np.log(img+1)/(np.log(1+np.max(img))))*255
#Specify the data type
img_log = np.array(img_log,dtype=np.uint8)

#Display the image

cv2_imshow(img)
cv2_imshow(img_log)
```

```python
image = cv2.imread('/content/drive/MyDrive/ColabPhotos/bluredText.png')

print("Normal Image: ")
cv2_imshow(image)


#Edge
edge_detection = np.array([[-1,-1,-1], 
                        [-1,8,-1], 
                        [-1,-1,-1]])
#Sharp
kernel_sharpening = np.array([[-1,-1,-1], 
                              [-1,9,-1], 
                              [-1,-1,-1]])
#Sharpen Kernel
sharpen_kernel = np.array([[0,-1,0], 
                         [-1,5,-1], 
                        [0,-1,0]])


# applying different kernels to the input image

sharpened = cv2.filter2D(image, -1, kernel_sharpening)
sharpened2 = cv2.filter2D(image, -1, sharpen_kernel)
sharpened3 = cv2.filter2D(image, -1, edge_detection)



print("Kernel Edges -1 Midle 9: ")
cv2_imshow(sharpened)

print("Kernels 0 -1 0\n       -1 5 -1 \n        0 -1 0:")
cv2_imshow(sharpened2)

print("Edge Detection")
cv2_imshow(sharpened3)

# I think best result is the first one edges is -1 middle is 9
```

```python
#8-16-24 and 100 pixel Image convertion
input = cv2.imread('/content/drive/MyDrive/ColabPhotos/grayscale.jpg')
print("Normal Size Image: ")
cv2_imshow(input)
#Get input size
 
height, width = input.shape[:2]

#Desired "pixelated" size
w, h = (8, 8)
 
#Resize input to "pixelated" size
temp = cv2.resize(input, (w, h))

#Initialize output image 8 bit
bit8 = cv2.resize(temp, (width, height), interpolation=cv2.INTER_NEAREST)
print("8-bit Image: ")
cv2_imshow(bit8)
#-----------------------------------------------------------------------

#Desired "pixelated" size
w, h = (16, 16)
 
#Resize input to "pixelated" size
temp = cv2.resize(input, (w, h))

#Initialize output image 16 bit
bit16 = cv2.resize(temp, (width, height), interpolation=cv2.INTER_NEAREST)
print("16-bit Image: ")
cv2_imshow(bit16)
#-----------------------------------------------------------------------

#Desired "pixelated" size
w, h = (24, 24)
 
#Resize input to "pixelated" size
temp = cv2.resize(input, (w, h))

#Initialize output image 24 bit
bit24 = cv2.resize(temp, (width, height), interpolation=cv2.INTER_NEAREST)
print("24-bit Image: ")
cv2_imshow(bit24)
#-----------------------------------------------------------------------

#Desired "pixelated" size
w, h = (100, 100)
 
#Resize input to "pixelated" size
temp = cv2.resize(input, (w, h))

#Initialize output image 100 bit
bit100 = cv2.resize(temp, (width, height), interpolation=cv2.INTER_NEAREST)
print("100-bit Image: ")
cv2_imshow(bit100)
#-----------------------------------------------------------------------
```

```python
from matplotlib import pyplot as plt

img0 = cv2.imread('/content/drive/MyDrive/ColabPhotos/NewYork.jpg')

# converting to gray scale
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)
print("Normal Image: ")
cv2_imshow(gray)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)
print("GaussianBlur Image: ")
cv2_imshow(img)

# Laplacian
laplacian = cv2.Laplacian(img,cv2.CV_64F)
print("Laplacian Image: ")
cv2_imshow(laplacian)

# Sobel
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  # x
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)  # y
print("SobelX Image: ")
cv2_imshow(sobelx)
print("SobelY Image: ")
cv2_imshow(sobely)

# Prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
prewittx = cv2.filter2D(img, -1, kernelx)
prewitty = cv2.filter2D(img, -1, kernely)
print("PrewittX Image: ")
cv2_imshow(prewittx)
print("PrewittY Image: ")
cv2_imshow(prewitty)


#Roberts Cross
from scipy import ndimage
robertsCrossX = np.array( [[1 , 0 ],
                          [ 0 ,-1 ]] )
  
robertsCrossY = np.array( [[ 0, 1 ],
                          [ -1, 0 ]] )
  

img = img / 255
vertical = ndimage.convolve( img, robertsCrossX )
horizontal = ndimage.convolve( img, robertsCrossY )
  
edged_img = np.sqrt( np.square(horizontal) + np.square(vertical))
edged_img = edged_img * 255
print("Roberts Cross Image: ")
cv2_imshow(edged_img)
```

```python
#Salt and Paper
import random
import skimage.io
from PIL import Image

def saltAndPaper(img):
  
    row , col = img.shape
     
    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
       
        # Pick a random y coordinate
        y_coord=random.randint(0, row - 1)
         
        # Pick a random x coordinate
        x_coord=random.randint(0, col - 1)
         
        # Color that pixel to white
        img[y_coord][x_coord] = 255
         
    
    number_of_pixels = random.randint(300 , 10000)
    for i in range(number_of_pixels):
       
        # Pick a random y coordinate
        y_coord=random.randint(0, row - 1)
         
        # Pick a random x coordinate
        x_coord=random.randint(0, col - 1)
         
        # Color that pixel to black
        img[y_coord][x_coord] = 0
         
    return img
 
img = cv2.imread('/content/drive/MyDrive/ColabPhotos/tree.jpg',0)
print("Normal Image: ")
cv2_imshow(img)


saltPaperImage = saltAndPaper(img)
print("Salt amd Paper Image Image: ")
cv2_imshow(saltPaperImage)



cv2.imwrite('/content/drive/MyDrive/ColabPhotos/saltAndPaper.jpg',saltPaperImage)


image = skimage.io.imread('/content/drive/MyDrive/ColabPhotos/saltAndPaper.jpg', as_gray=True)


#Normal Image Histogram
print("Normal Image Histogram")
hist,bins = np.histogram(img.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()



#Salt and Paper Image Histogram
print("Salt and Paper Image Histogram")
hist,bins = np.histogram(image.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(image.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```
