```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np


image = cv2.imread('face.jpg')

kernel1 = np.ones((5,5),np.float32)/30

img = cv2.filter2D(src=image, ddepth=-1, kernel=kernel1)

ksize= (10,10)
print("Normal Image:")
cv2_imshow(image)
print("Kernel Image:")
cv2_imshow(img)
print("Blur Image:")
blurImage = cv2.blur(image,ksize)
cv2_imshow(blurImage)
```

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np


imgBrain = cv2.imread('brainBlack.png')
median = cv2.medianBlur(imgBrain,5)
compare = np.concatenate((imgBrain,median),axis=1)

cv2_imshow(compare)
```
